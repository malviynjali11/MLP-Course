```python
from sklearn.pipeline import Pipeline
```

```python
st = [("Impute_mode", SimpleImputer(missing_values=np.nan, strategy="most_frequent")), 
      ("OHE", OneHotEncoder(sparse_output=False)), 
      ("Scaling", StandardScaler())]

pipe = Pipeline(steps = st).set_output(transform="pandas")
pipe
```

```python
X_train.head()
```

```python
pipe.fit_transform(X_train[["Fuel_Type", "Seller_Type"]])
```

```python
from sklearn.compose import ColumnTransformer

imputing_trf = [("median", SimpleImputer(missing_values=np.nan, strategy='median'), ['Kms_Driven']),
       ("mode", SimpleImputer(missing_values=np.nan, strategy='most_frequent'), ['Fuel_Type']),
       ("mean", SimpleImputer(missing_values=np.nan, strategy='mean'), ['Owner'])]


imputing_cols = ColumnTransformer(transformers=trf, remainder="passthrough", verbose_feature_names_out=False).set_output(transform="pandas")
imputing_cols
```

```python
df2 = imputing_cols.fit_transform(X_train)
```

```python
df2.head()
```

```python
X_train.head()
```

```python
encoding_trf = [("OHE", OneHotEncoder(sparse_output=False), ["Fuel_Type", "Seller_Type", "Transmission"]),
                ("OE", OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1), ["Car_Name"])]
encoded_col = ColumnTransformer(transformers=encoding_trf, remainder="passthrough").set_output(transform="pandas")
encoded_col
```

```python
df3 = encoded_col.fit_transform(X_train)
df3.head()
```

```python
df3.info()
```

```python
stp = [("Imputing", imputing_cols),
       ("Encoding", encoded_col),
       ("Scaling", StandardScaler())]

pipe = Pipeline(steps=stp).set_output(transform="pandas")
pipe
```

```python
df4 = pipe.fit_transform(X_train)
```

```python
from sklearn.feature_selection import RFE
from sklearn.svm import SVR

estimator = SVR(kernel="linear")
selector = RFE(estimator, n_features_to_select=5, step=1).set_output(transform="pandas")
```

```python
selector = selector.fit(df4, y_train)
selector.support_
```

```python
selector.ranking_
```

```python
selector.transform(df4)
```

```python
selector.estimator_.coef_
```

```python
estimator.fit(df4, y_train)
```

```python
stp = [("Imputing", imputing_cols),
       ("Encoding", encoded_col),
       ("Scaling", StandardScaler()),
       ("Feature Selection", selector)]

pipe = Pipeline(steps=stp).set_output(transform="pandas")
pipe
```
