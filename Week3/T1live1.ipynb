```python
from sklearn.compose import ColumnTransformer

trf = [("median", SimpleImputer(missing_values=np.nan, strategy='median'), ['Kms_Driven']),
       ("mode", SimpleImputer(missing_values=np.nan, strategy='most_frequent'), ['Fuel_Type']),
       ("mean", SimpleImputer(missing_values=np.nan, strategy='mean'), ['Owner'])]


ct = ColumnTransformer(transformers=trf, remainder="passthrough").set_output(transform="pandas")
ct
```

```python
df2 = ct.fit_transform(X_train)
```
